Question 2: What are the key assumptions of Simple Linear Regression?

Key Assumptions of Simple Linear Regression

Linearity
There is a linear relationship between the independent variable (X) and the dependent variable (Y).

Independence of Errors
The residuals (errors) are independent of each other; one observation does not influence another.

Homoscedasticity (Constant Variance)
The variance of the error terms is constant for all values of X.

Normality of Errors
The residuals are normally distributed, especially important for hypothesis testing.

Zero Mean of Errors
The average of the error terms is zero, meaning the model is unbiased.

ğŸ“Œ Why these assumptions matter

If these assumptions are violated, the regression estimates, predictions, and statistical tests may become unreliable.

âœ… In short:

Simple Linear Regression assumes a linear relationship, independent and normally distributed errors, and constant variance of errors.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)

Below is an example of Python code using scikit-learn to fit a Simple Linear Regression model and print the slope and intercept.

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (independent variable X and dependent variable y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Hours studied
y = np.array([2, 4, 5, 4, 5])                  # Exam scores

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Î²1):", model.coef_[0])
print("Intercept (Î²0):", model.intercept_)
```

Explanation

Slope (Î²â‚) shows how much the dependent variable changes for a one-unit increase in X.

Intercept (Î²â‚€) is the value of Y when X = 0.
